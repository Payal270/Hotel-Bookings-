**ROOM PRICE ANALYSIS**

```python
data['adr_updated']=data['adr']/(data['adults']+data['children'])
```

```python
resort = data[(data['hotel']=='Resort Hotel') & (data['is_canceled']==0)]
city = data[(data['hotel']=='City Hotel') & (data['is_canceled']==0)]
```

```python
resort_hotel=resort.groupby(['arrival_date_month'])['adr_updated'].mean().reset_index()
```

```python
resort_hotel
```

```python
city_hotel=city.groupby(['arrival_date_month'])['adr_updated'].mean().reset_index()
```

```python
final=resort_hotel.merge(city_hotel,on='arrival_date_month')
```

```python
final
```

```python
final.columns=['month','price_for_resort','price_for_city_hotel']
```

**HOW LONG DO PEOPLE STAY AT THE HOTELS?**

```python
valid_guests['total_nights']=valid_guests['stays_in_weekend_nights'] + valid_guests['stays_in_week_nights']
```

```python
valid_guests
```

```python
df1=valid_guests[['total_nights','hotel']]
```

```python
df1
```

```python
hotel_stay=df1.groupby(['total_nights','hotel']).value_counts().reset_index()
```

```python
hotel_stay
```

```python
hotel_stay.columns=['total_nights','hotel','number_of_stays']
```

```python
hotel_stay
```

```python

sns.catplot(x='total_nights', y='number_of_stays', kind='bar', hue='hotel', data=hotel_stay,height=10, aspect=1.5)

plt.title("Bar Chart of Total Nights vs. Number of Stays")
plt.xlabel("Total Nights")
plt.ylabel("Number of Stays")

plt.show()
```

It can be inferred that guests who stay for one week or less tend to choose city hotels, whereas those who prefer longer stays are more likely to opt for resorts. This conclusion is supported by the fact that very few guests choose to extend their visits beyond a week.

**ANALYSIS OF BOOKINGS BY MARKET SEGMENT**

```python
segments
```

**DISTRIBUTION OF PRICES ACROSS VARIOUS MARKET SEGMENTS** 

```python
plt.figure(figsize=(14,8))
sns.barplot(x='market_segment', y='adr_updated', hue='reserved_room_type',data=data)

plt.title('ADR by market segment and room type',fontsize=16)
plt.xlabel('Market segment',fontsize=14)
plt.ylabel('ADR per person',fontsize=14)
plt.xticks(rotation=45)
plt.show()

```

Based on our analysis, it appears that complementary bookings are the least expensive while bookings made through aviation tend to be more costly. One notable observation is that only Room Type L can be booked directly.
